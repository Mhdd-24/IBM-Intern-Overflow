#Data Augmentation

```python
# Import necessary lib.

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Data augmentation on training variable

train_datagen = ImageDataGenerator(rescale=1./255,
                                   zoom_range=0.2,
                                   horizontal_flip=True)
```

```python
# Data augmentation on testing variable

test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# Data augmentation on training data

xtrain = train_datagen.flow_from_directory('Flower_zip/Training',
                                           target_size=(64,64),
                                           class_mode='categorical',
                                           batch_size=100)
```

```python
# Data augmentation on testing data

xtest = test_datagen.flow_from_directory('Flower_zip/Testing',
                                         target_size=(64,64),
                                         class_mode='categorical',
                                         batch_size=100)
```

# CNN model training

```python
# Importing req. lib.

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D, MaxPooling2D, Flatten, Dense
```

```python
# Build a CNN block

model = Sequential() # Initializing sequential model
model.add(Convolution2D(32,(3,3),activation='relu',input_shape=(64,64,3))) # convolution layer
model.add(MaxPooling2D(pool_size=(2, 2))) # Max pooling layer
model.add(Flatten()) # Flatten layer
model.add(Dense(300,activation='relu')) # Hidden layer 1
model.add(Dense(150,activation='relu')) # Hidden layer 2
model.add(Dense(4,activation='softmax')) # Output layer
```

```python
# Compiling the model

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# Train model

model.fit_generator(xtrain,
                    steps_per_epoch=len(xtrain),
                    epochs=10,
                    validation_data=xtest,
                    validation_steps=len(xtest))
```

```python
# Save model

model.save('flower.h5')
```

# Model tuning

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
early_stop = EarlyStopping(monitor='val_accuracy', 
                           patience=5)

lr = ReduceLROnPlateau(monitor='val_accuaracy', 
                       factor=0.5, 
                       min_lr=0.00001)

callback = [early_stop,lr]
```

```python
# Train model

model.fit_generator(xtrain,
                    steps_per_epoch=len(xtrain),
                    epochs=100,
                    callbacks=callback,
                    validation_data=xtest,
                    validation_steps=len(xtest))
```
